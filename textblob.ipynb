```python
from textblob import TextBlob
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
text= open("/content/drive/MyDrive/Colab Notebooks/nlp/openfile.txt")
text=text.read()
```

```python
blob=TextBlob(text)
```

```python
words=blob.words
print(words)
```

```python
tags=blob.tags
print(tags)
```

```python
sentences=blob.sentences
print(sentences)
```

```python
noun_phrase=blob.noun_phrases
print(noun_phrase)
```

```python
np_count=blob.np_counts
print(np_count)
```

```python
polarity=blob.polarity
print(polarity)
```

```python
subjectivity=blob.subjectivity
print(subjectivity)
```

```python
sentiment=blob.sentiment
print(sentiment)
```
