## list comprehension

```python
x = [1,2,3,4]
```

```python
out = []
for item in x:
    out.append(item**2)
print(out)
```

```python
[item**2 for item in x]
```

## map and filter

```python
seq = [1,2,3,4,5]
```

```python
map(times2,seq)
```

```python
list(map(times2,seq))
```

```python
list(map(lambda var: var*2,seq))
```

```python
filter(lambda item: item%2 == 0,seq)
```

```python
list(filter(lambda item: item%2 == 0,seq))
```
